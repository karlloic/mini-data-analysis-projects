# NYC Taxi Trips - When should I hit the road in NYC

The purpose of this is to attempt to determine which is the best time to travel (by car) through the City of New York. This project came about when I was planning on traveling to NYC and deciding which flight time to pick. The ultimate goal was for me to arrive and be able to hit the road when there is least traffic.

As you will see in the plot below, there is almost always heavy traffic in NYC except for 5am when car speeds reach ~20mph and also late at night. I therefore decided to land between 5 and 6 am.

```python
import pandas as pd
import requests
import StringIO
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series, DataFrame

%matplotlib inline

def debug(f, *args, **kwargs):
    from IPython.core.debugger import Pdb 
    pdb = Pdb(color_scheme='Linux')
    return pdb.runcall(f, *args, **kwargs)
```

```python
def getTripData(year):
    """
    Return a merged DataFrame containing the income, 
    country name and region for a given year. 

    Parameters:
        year (int): The year of interest

    Returns
    -------
    a DataFrame
       A pandas DataFrame with with Trip Data. Data is described here
       Green -> http://www.nyc.gov/html/tlc/downloads/pdf/data_dictionary_trip_records_green.pdf
       Yellow -> http://www.nyc.gov/html/tlc/downloads/pdf/data_dictionary_trip_records_yellow.pdf

    Example
    -------
    >>> getTripData(2010)
    """
    dateColumns = ['Lpep_dropoff_datetime', 'lpep_pickup_datetime']
    #Attempt to load previous backed up file
    try:
        print 'Attempting to load local backup file...'
        trips = pd.read_csv('data/%s_nyc_taxi_trips_backup.csv' %(str(year)), parse_dates=dateColumns)
    except:
        print 'Failed to load backup file; not found'
        trips = pd.DataFrame()
        taxiColor = ('yellow', 'green')
        for tc in taxiColor:
            for month in range(1, 13): # Remember we are only retrieving data from one month
                url = 'https://storage.googleapis.com/tlc-trip-data/%d/%s_tripdata_%d-%s.csv' %(year, tc, year, str(month).zfill(2))
                print 'Retrieving data from TLC website... %s' %(url)
                r = requests.get(url)
                if r.status_code == requests.codes.ok:
                    try:
                        lines = StringIO.StringIO(r.content).readlines()
                    except:
                        continue
                else:
                    print 'No %d/%d data for %s taxis' %(month.zfill(2), year, tc)
                    continue
                nl = ''
                for l in lines:
                    nl += l[:-4] + '\r\n' # CSV file is malformated; removing two extra commas at end of lines
                #ns = StringIO.StringIO(nl)
                if trips.shape[0] == 0:
                    trips = pd.read_csv(StringIO.StringIO(nl), parse_dates=dateColumns)
                else:
                    trips = trips.append(pd.read_csv(StringIO.StringIO(nl)), ignore_index=True, , parse_dates=dateColumns)
        trips.to_csv('data/%s_nyc_taxi_trips_backup.csv' %(str(year)))
        print 'Successfully loaded data and backed up locally'
        print trips.shape
    print 'Data successfully loaded'
    return trips
```

```python
trips = getTripData(2015)
```

```python
# Evaluating total time traveled in a new column in the dataframe (in hours)
trips['Trip_duration'] = trips['Lpep_dropoff_datetime'] - trips['lpep_pickup_datetime']
trips['Trip_duration'] = trips['Trip_duration'].dt.total_seconds() / 3600
```

```python
# Verify new column data
trips[['Trip_distance', 'Trip_duration']].head()
```

```python
# Get average trip speed in mph
trips['Trip_speed'] = trips['Trip_distance'] / trips['Trip_duration']
```

```python
# Create new data frame with only valuable data
trips = trips[['Trip_speed', 'lpep_pickup_datetime']].copy()
trips.head()
```

```python
# Sort the resulting dataframe by pickup time
trips = trips.sort_values(by='lpep_pickup_datetime')
```

## <font color='red'>Removing data resulting due Human error or Device malfunctioning</font>

### Discovery
After plotting the data for the data the first time, I noticed some spikes in the Taxi Speed. Some of the speeds went up to 200,000 mph. Based on that, I assumed that there must have been some odd data. After exploring the dataset again, I detected some strange entries. For example if you interpret the row below:
                pickup_time             dropoff_time      distance
    1432692	| 2015-09-29 17:01:17 |	2015-09-29 17:01:19 | 112.60
    
it means this cab traveled 112.6 miles in 2 seconds.
    
### Conclusion
It can therefore be concluded that there must have been some device or user error. 

### Solution
Here are the actions taken to attempt to correct the problem found above
    1. Drop all NaNs
    2. Cap the cab speed i.e. Remove all rows with Trip_speed  5 mph above the city speed limit (25mph)
        http://www.nyc.gov/html/dot/downloads/pdf/2014-10-twenty-five-mile-speed-limit-faq.pdf
    3. Set minimum speed to 1 mph

```python
# Remove any rows with NaNs and values above city speed limit
trips = trips[trips['Trip_speed'] < 30]
trips = trips[trips['Trip_speed'] > 1]
trips.dropna(inplace=True)
```

```python
# Since matplotlib cannot plot millions of data points,
# reduce the number of data points by averaging the times
trips['lpep_pickup_datetime'] = trips['lpep_pickup_datetime'].dt.time
trips = trips.groupby(by='lpep_pickup_datetime').mean()
trips = trips.reset_index()
```

```python
# Plot of average Taxi speeds over time
plt.figure(figsize=(50,20))
plt.plot(trips['lpep_pickup_datetime'], trips['Trip_speed'], 'y-')
plt.title("\nNYC Taxi trips\n", fontsize=40)
plt.ylabel("Taxi speed (mph)", fontsize=30)
plt.xlabel('Time of day', fontsize=30)
plt.yticks(fontsize=25)
plt.xticks(fontsize=25)
plt.show()
```
